```python
import pandas as pd

# Load the data again to ensure all original columns are present
df = pd.read_csv('/content/Psychosocial_Health_Analysis.csv')
```

```python
# Count the occurrences of each problem category
problem_category_counts = df['problem_category'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=problem_category_counts.index, y=problem_category_counts.values, palette='viridis')
plt.title('Distribution of Problem Categories')
plt.xlabel('Problem Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Filter the DataFrame for 'Relationship' problem category and age between 20 and 25
relationship_problems_df = df[(df['problem_category'] == 'Relationship') & (df['Age'] >= 20) & (df['Age'] <= 25)]

# Count the occurrences of each unique problem description within the filtered data
problem_description_counts = relationship_problems_df['Problem_description'].value_counts()

# Display the results
print("Count of each Problem Description in 'Relationship' Category for age 20-25:")
display(problem_description_counts)
```

```python
# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(problem_description_counts, labels=problem_description_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Problem Descriptions in "Relationship" Category (Age 20-25)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
# Filter the DataFrame for 'Relationship' problem category and age above 25
relationship_problems_above_25_df = df[(df['problem_category'] == 'Relationship') & (df['Age'] > 25)]

# Display the problem descriptions
print("Problem Descriptions for 'Relationship' Category (Age > 25):")
for description in relationship_problems_above_25_df['Problem_description']:
    print(f"- {description}\n")
```
